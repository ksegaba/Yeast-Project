## PCA of isolate fitness values across environments

```python
df = pheno.transpose()#.stack().reset_index()
#df.columns = ["Condition", "ID", "Fitness"]
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Standardizing the features
x = StandardScaler().fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
df_pca = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
df_pca = pd.concat([df_pca, pd.Series(df.index)], axis = 1)
df_pca.head()
```

```python
import seaborn as sns
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
sns.scatterplot(data=df_pca, x="principal component 1", y="principal component 2", hue=0, legend=False)
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.title('Conditions PCA', fontsize = 20)
for i in range(df_pca.shape[0]):
    plt.text(x=df_pca["principal component 1"][i]+0.3, y=df_pca["principal component 2"][i]+0.3, 
    s=df_pca[0][i], fontdict=dict(size=7))
```

## Fitness vs. Kinship

```python
import pandas as pd
import datatable as dt
kinship = dt.fread("/mnt/home/seguraab/Shiu_Lab/Project/Data/Peter_2018/geno_transposed.csv_ordered_kinship.txt")
kinship = kinship.to_pandas()
cols = kinship['C0'].tolist()
kinship.set_index('C0', inplace=True)
kinship.columns = cols
kinship.head()
```

```python
k_stack = kinship.stack().reset_index()
k_stack.columns = ['ID_1','ID_2','Kinship']
k_stack = k_stack[k_stack["ID_1"] != k_stack["ID_2"]]
k_stack.head()
```

```python
pheno = pd.read_csv("/mnt/home/seguraab/Shiu_Lab/Project/Data/Peter_2018/pheno.csv")
pheno.set_index('Unnamed: 0', inplace=True)
pheno = pheno.transpose().corr(method="pearson") # isolate correlation across isolates
pheno.head()
```

```python
pheno = pheno.rename_axis(None).rename_axis(None, axis=1)
p_stack = pheno.stack().reset_index()
p_stack.columns = ["ID_1", "ID_2", "PCC"]
p_stack = p_stack[p_stack["ID_1"] != p_stack["ID_2"]]
p_stack.head()
```

```python
k_stack.shape, p_stack.shape
```

```python
k_stack.Kinship.min(), k_stack.Kinship.max(), p_stack.PCC.min(), p_stack.PCC.max()
```

```python
df = pd.merge(k_stack, p_stack,  how='left', left_on=['ID_1','ID_2'], right_on = ['ID_1','ID_2'])
df.head(), df.shape
```

```python
import matplotlib.pyplot as plt
import mpl_scatter_density
from matplotlib.colors import LinearSegmentedColormap

# "Viridis-like" colormap with white background
white_viridis = LinearSegmentedColormap.from_list('white_viridis', [
    (0, '#ffffff'),
    (1e-20, '#440053'),
    (0.2, '#404388'),
    (0.4, '#2a788e'),
    (0.6, '#21a784'),
    (0.8, '#78d151'),
    (1, '#fde624'),
], N=256)

RdBu = LinearSegmentedColormap.from_list('RdBu', [
    (0, '#ffffff'),
    (1e-20, '#67001f'),
    (0.2, '#ca4842'),
    (0.4, '#f7b799'),
    (0.6, '#a7d0e4'),
    (0.8, '#3884bb'),
    (1, '#053061'),
], N=256)

def using_mpl_scatter_density(fig, x, y):
    ax = fig.add_subplot(1, 1, 1, projection='scatter_density')
    density = ax.scatter_density(x, y, cmap=RdBu)
    plt.xlabel("Kinship")
    plt.ylabel("Fitness Correlation Across Isolates")
    fig.colorbar(density, label='Number of points per pixel')

fig = plt.figure()
using_mpl_scatter_density(fig, k_stack.Kinship, p_stack.PCC)

plt.tight_layout()
plt.savefig("kinship_vs_phenoIsoCorr.pdf", bbox_inches="tight")
plt.show()
```
