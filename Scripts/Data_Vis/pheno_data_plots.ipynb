# Peter 2018 phenotype data plots

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pheno = pd.read_csv("/mnt/home/seguraab/Shiu_Lab/Project/Data/Peter_2018/pheno.csv")
pheno.head()
```

```python
pheno.set_index("Unnamed: 0", inplace=True)
pheno.describe()
```

## Heatmap of fitness values for each isolate across environments

```python
fig=sns.clustermap(pheno, figsize=(12,30), cmap="RdBu_r", yticklabels=False, dendrogram_ratio=(.1, .1)) # cbar_pos=(1, .7, 0.02, .2),
ax = fig.ax_heatmap
ax.set_xlabel("Environments",    )
ax.set_ylabel("Isolates")
ax.tick_params(right=False)
plt.tight_layout(rect=[0, 0.7, 0, 0.7]) # left, bottom, right, top
plt.show(ax)
plt.savefig("/mnt/home/seguraab/Shiu_Lab/Project/Scripts/Data_Vis/pheno_heatmap.pdf")
```

## Correlation heatmap of fitness values
1. Comparing environments
2. Comparing isolates

```python
pheno_corr_envs = pheno.corr(method="pearson")
pheno_corr_envs.head()
```

```python
pheno_corr_isolates = pheno.transpose().corr(method="pearson")
pheno_corr_isolates.head()
```

```python
fig = sns.clustermap(pheno_corr_envs, figsize=(10,10), cmap="RdBu_r", center=0, dendrogram_ratio=(.1, .1)) #cbar_pos=(1, .7, 0.02, .2),
ax = fig.ax_heatmap
ax.set_xlabel("Environments")#, font_size=14)#, fontweight="bold")
ax.set_ylabel("Environments")#, font_size=14)#, fontweight="bold")
plt.tight_layout(rect=[0.5, 0.5, 0.5, 0.5]) # left, bottom, right, top
plt.savefig("/mnt/home/seguraab/Shiu_Lab/Project/Scripts/Data_Vis/pheno_envs_corr.pdf", bbox_inches = "tight")
```

```python
fig = sns.clustermap(pheno_corr_isolates, figsize=(12,12), cmap="RdBu_r", yticklabels=False, xticklabels=False, center=0, dendrogram_ratio=(.1, .1)) #cbar_pos=(, .7, 0.02, .2), 
ax = fig.ax_heatmap
ax.set_xlabel("Isolates")
ax.set_ylabel("Isolates")
ax.tick_params(right=False, bottom=False)
plt.tight_layout(rect=[0, 0.7, 0, 0.7]) # left, bottom, right, top
plt.savefig("/mnt/home/seguraab/Shiu_Lab/Project/Scripts/Data_Vis/pheno_isolates_corr.pdf", bbox_inches = "tight")
```
