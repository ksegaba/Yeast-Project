## PCA of isolate fitness values across environments

```python
df = pheno.transpose()#.stack().reset_index()
#df.columns = ["Condition", "ID", "Fitness"]
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Standardizing the features
x = StandardScaler().fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
df_pca = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
df_pca = pd.concat([df_pca, pd.Series(df.index)], axis = 1)
df_pca.head()
```

```python
import seaborn as sns
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
sns.scatterplot(data=df_pca, x="principal component 1", y="principal component 2", hue=0, legend=False)
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.title('Conditions PCA', fontsize = 20)
for i in range(df_pca.shape[0]):
    plt.text(x=df_pca["principal component 1"][i]+0.3, y=df_pca["principal component 2"][i]+0.3, 
    s=df_pca[0][i], fontdict=dict(size=7))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# Standardizing the features
x = StandardScaler().fit_transform(pheno)

# PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
pheno_pca = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
pheno_pca = pd.concat([pheno_pca, pd.Series(pheno.index)], axis = 1)
pheno_pca.head()
```

```python
import seaborn as sns
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
sns.scatterplot(data=pheno_pca, x="principal component 1", y="principal component 2", hue=0, legend=False)
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.title('Conditions PCA', fontsize = 20)
#for i in range(pheno_pca.shape[0]):
#    plt.text(x=pheno_pca["principal component 1"][i]+0.3, y=pheno_pca["principal component 2"][i]+0.3, 
#     s=pheno_pca["Unnamed: 0"][i], fontdict=dict(size=7))
```

## Fitness vs. Kinship

```python
import pandas as pd
import datatable as dt
kinship = dt.fread("/mnt/home/seguraab/Shiu_Lab/Project/Data/Peter_2018/geno_transposed.csv_ordered_kinship.txt")
kinship = kinship.to_pandas()
cols = kinship['C0'].tolist()
kinship.set_index('C0', inplace=True)
kinship.columns = cols
kinship.head()
```

```python
kinship.stack = kinship.stack().reset_index()
kinship.stack.columns = ['ID_1','ID_2','Kinship']
kinship.stack = kinship.stack[kinship.stack["ID_1"] != kinship.stack["ID_2"]]
kinship.stack.head()
```

```python
pheno = pd.read_csv("/mnt/home/seguraab/Shiu_Lab/Project/Data/Peter_2018/pheno.csv")
pheno.set_index('Unnamed: 0', inplace=True)
pheno = pheno.transpose().corr(method="pearson") # isolate correlation across envs
pheno.head()
```

```python
pheno = pheno.rename_axis(None).rename_axis(None, axis=1)
pheno.stack = pheno.stack().reset_index()
pheno.stack.columns = ["ID_1", "ID_2", "PCC"]
pheno.stack = pheno.stack[pheno.stack["ID_1"] != pheno.stack["ID_2"]]
pheno.stack.head()
```

```python
import mpl_scatter_density
from matplotlib.colors import LinearSegmentedColormap

# "Viridis-like" colormap with white background
white_viridis = LinearSegmentedColormap.from_list('white_viridis', [
    (0, '#ffffff'),
    (1e-20, '#440053'),
    (0.2, '#404388'),
    (0.4, '#2a788e'),
    (0.6, '#21a784'),
    (0.8, '#78d151'),
    (1, '#fde624'),
], N=256)

def using_mpl_scatter_density(fig, x, y):
    ax = fig.add_subplot(1, 1, 1, projection='scatter_density')
    density = ax.scatter_density(x, y, cmap=white_viridis)
    fig.colorbar(density, label='Number of points per pixel')

fig = plt.figure()
using_mpl_scatter_density(fig, kinship.stack.Kinship, pheno.stack.PCC)
plt.show()
```
